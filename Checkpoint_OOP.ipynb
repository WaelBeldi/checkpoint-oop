```python
class Account:
  def __init__(self, account_number, account_balance, account_holder):
    self.account_number = account_number
    self.account_balance = account_balance
    self.account_holder = account_holder

  def deposit(self, amount):
    self.account_balance += amount

  def withdraw(self, amount):
    if self.account_balance >= amount:
      self.account_balance -= amount
    else:
      print("Insufficient funds")

  def check_balance(self):
    print(f"Account balance for {self.account_holder}: ${self.account_balance}")

my_account = Account("123456789", 1000, "Wael")
my_account.deposit(500)
my_account.withdraw(200)
my_account.check_balance()

account1 = Account("111111111", 1200, "Yassine")
account1.deposit(500)
account1.withdraw(300)
account1.check_balance()

account2 = Account("222222222", 1500, "Amine")
account2.deposit(700)
account2.withdraw(400)
account2.check_balance()
```
